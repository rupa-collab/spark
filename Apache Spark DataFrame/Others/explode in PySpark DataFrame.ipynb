#### Using explode() in DataFrame
  
explode() is used to flatten arrays or maps — it converts each element of an array/map into a separate row.



##### With Arrays

```python
from pyspark.sql.functions import explode
from pyspark.sql import SparkSession

data = [("Alice", ["Python", "Scala", "SQL"]),
        ("Bob",   ["Java", "R"])]

df = spark.createDataFrame(data, ["name", "skills"])
df.show()

df.select("name", explode("skills").alias("skill")).show()
```

```python
data = [("Alice", {"math": 90, "science": 85})]
df = spark.createDataFrame(data, ["name", "scores"])

df.select("name", explode("scores").alias("subject", "marks")).show()
```

##### Variants

```python
from pyspark.sql.functions import posexplode
data = [("Alice", ["Python", "Scala", "SQL"]),
        ("Bob",   ["Java", "R"])]

df = spark.createDataFrame(data, ["name", "skills"])


df.select("name", posexplode("skills").alias("pos", "skill")).show()
# pos=0 → Python, pos=1 → Scala, pos=2 → SQL
```

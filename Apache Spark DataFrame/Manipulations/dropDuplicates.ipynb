**df.dropDuplicates(["name", "country"]) — In Detail
**
It drops duplicate rows based only on the specified columns (name and country). For rows where these two columns have identical values, Spark keeps one row and drops the rest.

```python
data = [
    ("Alice", "IN", 30, "Engineer"),
    ("Alice", "IN", 25, "Manager"),    # same name+country as row 1
    ("Alice", "US", 30, "Engineer"),   # same name, different country
    ("Bob",   "IN", 40, "Analyst"),
    ("Bob",   "IN", 40, "Analyst"),    # exact duplicate of row 4
]
df = spark.createDataFrame(data, ["name", "country", "age", "role"])
df.show()
```

Spark keeps one row per unique (name, country) combination — the other columns (age, role) don't matter.



```python
df.dropDuplicates(["name", "country"]).show()
```

Comparison with distinct()


```python
df.distinct().show()                           # ALL columns must match to be a duplicate
df.dropDuplicates().show()                      # same as distinct()
df.dropDuplicates(["name", "country"]).show()   # only these columns define a duplicate
```

### Renaming Columns

#### Summary
Best choice depends on use case:

- All columns → toDF
- Subset of columns → select with alias
- Dynamic mapping → reduce with withColumnRenamed

```python
# Create sample data and initialize DataFrame with column names
data = [("John", 20, "US"),
        ("Kay", 32, "US"),
        ("Harry", 24, "IN"),
        ("David", 30, "EU")]

df = spark.createDataFrame(data, ("name", "age", "country"))
df.show()  # Display the original DataFrame
```

#### 1. Chaining withColumnRenamed

```python
df.withColumnRenamed("name", "Employee Name") \
    .withColumnRenamed("age", "Employee Age") \
    .withColumnRenamed("country", "Employee Country") \
    .show()

```

#### 2. toDF — rename all columns at once

```python
renamed_df = df.toDF("Employee Name", "Employee Age", "Employee Country")
display(renamed_df)
```

#### 3. Loop with reduce - for dynamic column renaming

```python
from functools import reduce

rename_map = {"Employee Name": "Name", "Employee Country": "Country"}

df_renamed = reduce(
    lambda renamed_df, cols: renamed_df.withColumnRenamed(*cols),
    rename_map.items(),
    renamed_df
)
display(df_renamed)
```

```python
df.show()
```

#### 4. select with aliases (rename specific columns)

```python
from pyspark.sql.functions import col
df.select(col("name").alias("Name"), 
          col("age").alias("Age"),
          col("country").alias("Country")).show()
```

#### 5. selectExpr


```python
df.selectExpr("name as Name", "age as Age", "country as Country").show()
```

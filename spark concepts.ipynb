A DataFrame wraps data with a named, typed schema:

root
 |-- name:    string (nullable = true)
 |-- age:     integer (nullable = true)
 |-- country: string (nullable = true)

- Catalyst optimizer can use schema info to optimize query plans
- Full SQL support via spark.sql("SELECT * FROM ...")



DataFrame
├── Schema (StructType)     ← metadata only, lives in Driver
│     ├── name: String
│     ├── age: Integer
│     └── country: String
└── Underlying RDD[Row]     ← actual data distributed across executors

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([
    StructField("name",    StringType(),    True),
    StructField("age",     IntegerType(),   True),
    StructField("country", StringType(),    True)
])

df = spark.createDataFrame([("Alice", 30, "IN"), ("Bob", 25, "US")], schema)
df.printSchema()
```

```python
df1 = df.filter(df.age > 25)
```

```python
df1.show()
```

```python
df1.select("name"
           , "country"
           ).show()
```

```python
data = [
    ("Alice", 30),
    ("Bob",   None),   # age is null
    (None,    25),     # name is null
    (None,    None),   # both null
]
df = spark.createDataFrame(data, ["name", "age"])
df1 = df.fillna({"age":0, "name":"unknown"})
display(df1)
df1.replace("unknown", None, "name").show()
```

```python
data = [
    ("Alice", 30),
    ("Bob",   None),   # age is null
    (None,    25),     # name is null
    (None,    None),   # both null
]
df = spark.createDataFrame(data, ["name", "age"])
df.distinct().show()
```

```python
data = [
    ("Alice", "IN", 30, "Engineer"),
    ("Alice", "IN", 25, "Manager"),    # same name+country as row 1
    ("Alice", "US", 30, "Engineer"),   # same name, different country
    ("Bob",   "IN", 40, "Analyst"),
    ("Bob",   "IN", 40, "Analyst"),    # exact duplicate of row 4
]
df = spark.createDataFrame(data, ["name", "country", "age", "role"])
df.show()
```

```python
df.dropDuplicates(["name", "country"]).show()
```
